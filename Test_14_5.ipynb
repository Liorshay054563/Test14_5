```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
items_df = pd.read_csv('Items.csv')
display(items_df.head())
items_df.info()
items_df.describe().T # the answer olso here in mean - item_price
# 1
avg_price = items_df['item_price'].mean()
print('The avg price of items is - ',avg_price)
```

```python
### graph - 1 -Gender
customers_df = pd.read_csv('Customers.csv')
customers_df.info()
customers = customers_df.dropna()
display(customers.head())

gender_counts = customers['gender'].value_counts()
gender_counts.plot(kind='bar')

plt.title('Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
### graph - 2 - nationallity

nation_counts = customers['nationallity'].value_counts()
nation_counts.plot(kind='bar')

plt.title('Customers by Nations')
plt.xlabel('Nations')
plt.ylabel('Count')

plt.show()
```

```python
### graph - 3 - hist age

nation_counts = customers['age'].value_counts()
nation_counts.plot(kind='hist')

plt.title('Customers by age')
plt.xlabel('Years old')
plt.ylabel('Count')

plt.show()
```
